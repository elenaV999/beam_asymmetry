## Rotation matrix

4 rotation functions: 
- rot(v1,v2)  --> better for beam rotation but slower ($\sim 85 \mu s$)  
- rot(R1, v2) + optimized version  --> use opt version for full convolution ($\sim 5 \mu s$)

Verified that they differ by numerical errors of $10^{-17}$

```python
#%matplotlib widget
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
import timeit
import importlib

import functions_rot as frot
```

```python
## Set: initial and final point of the rotation  
phi1, theta1 = 30., 45.
phi2, theta2 = 15., 60.
vc1=hp.pixelfunc.ang2vec(phi1, theta1, lonlat=True)
vc2=hp.pixelfunc.ang2vec(phi2, theta2, lonlat=True)
print('vc1 = ', vc1 )
print('vc2 = ', vc2 )
```

```python
# Rotation matrix from initial and final position
Rot_matrix=frot.rotmatrix_frame_vec(vc1, vc2)  
print(Rot_matrix)
```

```python
# Rotation matrix to final position (build rotation to initial position in advance)
R1=frot.get_R1(vc1)
Rot_matrix=frot.rotmatrix_frame(R1, vc2)
print(Rot_matrix)
```

```python
# Rotation matrix to final position - same but optimized for final run
R1=frot.get_R1(vc1)
Rot_matrix=frot.rotmatrix_frame_opt(R1, vc2)
print(Rot_matrix)
```

```python
print('Vec rotation - optimized R1 rot\n', frot.rotmatrix_frame_vec(vc1, vc2)-frot.rotmatrix_frame_opt(R1, vc2), '\n' )
print('Vec rotation - R1 rot\n', frot.rotmatrix_frame_vec(vc1, vc2)-frot.rotmatrix_frame(R1, vc2), '\n'  )
print('oprimized R1 rot - R1 rot \n', frot.rotmatrix_frame_opt(R1, vc2)-frot.rotmatrix_frame(R1, vc2) )
```
